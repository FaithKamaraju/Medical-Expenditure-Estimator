```python
from sklearn import linear_model
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("expenses.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
cols = ['sex','smoker','region']
df[cols] = df[cols].apply(lambda x: pd.factorize(x)[0])
```

```python
df.head()
```

```python
X_df = df[['age', 'sex', 'bmi', 'children', 'smoker', 'region']]
Y_df = df['charges']
```

```python
normalized_X_df=(X_df-X_df.mean())/X_df.std()
```

```python
normalized_X_df.head()
```

```python
normalized_df = pd.concat([normalized_X_df,Y_df],axis=1)
```

```python
normalized_df
```

```python
train_df, test_df = train_test_split(normalized_df,test_size=0.2)
```

```python
print(len(train_df), len(test_df))
```

```python
reg_model = linear_model.Ridge()
```

```python
reg_model.fit(train_df[['age', 'sex', 'bmi', 'children', 'smoker', 'region',]],train_df['charges'],)
```

```python
reg_model.score(train_df[['age', 'sex', 'bmi', 'children', 'smoker', 'region',]],train_df['charges'],)
```

```python
reg_model.score(test_df[['age', 'sex', 'bmi', 'children', 'smoker', 'region',]],test_df['charges'])
```
